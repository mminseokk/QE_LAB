```python
df = pd.read_csv('setfi1.csv',na_values=' ')
X = df.iloc[:4000,1:301]
Y = df.iloc[:4000,-1]

data = pd.concat([X,Y],axis = 1)
```

```python
selected_feature_ensemble  = [14,24,132,146,197,206,300]
data_selected_ensemble = data.iloc[:,selected_feature_ensemble]
```

```python
model_ensemble = CTGAN()
model_ensemble.fit(data_selected_ensemble)
```

```python
tvae_ensemble = TVAE()
tvae_ensemble.fit(data_selected_ensemble)
```

```python
import pickle
with open('model_ctgan_small.pkl', 'wb') as f:
    pickle.dump(model_ensemble, f)

with open('model_tvae_small.pkl', 'wb') as f:
    pickle.dump(tvae_ensemble, f)
```
