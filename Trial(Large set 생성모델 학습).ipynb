```python
df = pd.read_csv('setfi1.csv',na_values=' ')
X = df.iloc[:4000,1:301]
Y = df.iloc[:4000,-1]

data = pd.concat([X,Y],axis = 1)
```

```python
from sdv.tabular import CTGAN
```

```python
from sdv.tabular import TVAE
```

```python
model_ctgan = CTGAN()
model_ctgan.fit(data)
```

```python
model_tvae = TVAE()
model_tvae.fit(data)
```

```python
synthetic_data_ctgan  = model_ctgan.sample(4000)
```

```python
synthetic_data_tvae  = model_tvae.sample(4000)
```

```python
#파일 저장
import pickle
with open('model_ctgan.pkl', 'wb') as f:
    pickle.dump(model_ctgan, f)

with open('model_tvae.pkl', 'wb') as f:
    pickle.dump(model_tvae, f)
```
