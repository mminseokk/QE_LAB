#5.데이터 정확도

##5.1 중복 제조 경로 갯수 확인

```python
duplication_1 = pd.merge(synthetic_data_ensemble_1, data_selected,
                          left_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          right_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          how='inner')
duplication_2 = pd.merge(synthetic_data_ensemble_2, data_selected,
                          left_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          right_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          how='inner')
duplication_3 = pd.merge(synthetic_data_ensemble_3, data_selected,
                          left_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          right_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          how='inner')
duplication_4 = pd.merge(synthetic_data_ensemble_4, data_selected,
                          left_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          right_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          how='inner')
duplication_5 = pd.merge(synthetic_data_ensemble_5, data_selected,
                          left_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          right_on=['locn15', 'locn25','locn133','locn147','locn198','locn207'],
                          how='inner')
```

```python
duplication_1.shape, duplication_2.shape, duplication_3.shape, duplication_4.shape, duplication_5.shape
```

##5.2 라벨 인코딩 정확도 확인

```python
List = [duplication_1, duplication_2, duplication_3, duplication_4, duplication_5]
for dataframe in List:
  dataframe.loc[dataframe['res_x'] <= limit, 'res_x'] = 0
  dataframe.loc[dataframe['res_x'] > limit, 'res_x'] = 1
  dataframe.loc[dataframe['res_y'] <= limit, 'res_y'] = 0
  dataframe.loc[dataframe['res_y'] > limit, 'res_y'] = 1
```

```python
from sklearn.metrics import accuracy_score
acc_list = []
for dataframe in List:
  acc = accuracy_score(dataframe['res_x'],dataframe['res_y'])
  acc_list.append(acc)
```
